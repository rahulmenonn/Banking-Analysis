```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Banking.csv')
df.head()
```

```python
df.describe()
```

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Medium','high']
df['Income Bands']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income Bands'].value_counts().plot(kind='bar')
```

```python
categorical_cols=df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Bands"]].columns

for col in categorical_cols:
  print(f"Value counts for '{col}':")
  display(df[col].value_counts())
```

Univariate Analysis

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Bands"]]):
                             plt.figure(i)
                             sns.countplot(data=df,x=predictor)
```

Bivariate analysis

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Bands"]]):
                             plt.figure(i)
                             sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
from IPython.core.pylabtools import figsize
#Histogram of value counts for different occupation

for col in categorical_cols:
  if col=="Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title("Histogram of occupation count")
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

##Numerical Analysis

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

#Univariate analysis and visualisation
plt.figure(figsize=(15,15))
for i,col in enumerate(numerical_cols):
  plt.subplot(3,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.tight_layout()
plt.show()
```

##Heatmaps

```python
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

##Insights of EDA:

The strongest possible  correlation occurs among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
